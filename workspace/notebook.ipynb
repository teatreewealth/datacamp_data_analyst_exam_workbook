# Recommended metric

Based on the observations from our analysis of the data during the first 6 weeks from launch, we saw that units sold drives the revenue increase across all sales approaches. In week 6, more than 50% of the total revenue is generated by Email + Call. This is the case even with a decreasing average unit price that goes from 15.67 in week 1 to 14.67 in week 6 for the same sales approach. **Based on these observations we recommend to monitor the success of the campain using the Average Items Sold**. Considering that the average number of items sold with Email + Call during the 6 weeks campaign was 15 and that was making more than 50% of the overall renueve, any number above 30 will indicate a sign of achieving the sales goal.    

# Recommendation

**The recommendation is to use Email + Call going forward. This is the method that generates higher revenue consistently over time and is quite easy to implement.**

- The evidence points at a moderate positive influence of website visits on revenue and units sold. We recommend to keep the website updated with the products information and latest prices.
